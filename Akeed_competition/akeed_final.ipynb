**Step 2:-**Other steps in data preprocessing from the data created in previous steps

```
train_df1   = train_vendor_data_merged
test_df    = test_vendor_data_merged
train_df1.target.value_counts()
```

```
reduce_mem_usage(train_df1)
reduce_mem_usage(test_df)
```

```
index_to_use_later = test_df['CID X LOC_NUM X VENDOR'].values
train_y = train_df1['target'].values
train_df1 = train_df1.drop(['target'],axis=1)
columns_to_drop = ['CID X LOC_NUM X VENDOR']
train_df1 = train_df1.drop(columns_to_drop,axis=1)
test_df  = test_df.drop(columns_to_drop,axis=1)
```

```
import gc
def encode_LE(train,test,cols,verbose=True):
    for col in cols:
        df_comb = pd.concat([train[col],test[col]],axis=0)
        df_comb,_ = df_comb.factorize(sort=True)
        nm = col
        if df_comb.max()>32000: 
            train[nm] = df_comb[:len(train)].astype('int32')
            test[nm] = df_comb[len(train):].astype('int32')
        else:
            train[nm] = df_comb[:len(train)].astype('int16')
            test[nm] = df_comb[len(train):].astype('int16')
        del df_comb; x=gc.collect()
        if verbose: print(nm,', ',end='')

encode_LE(train_df1, test_df,['id','vendor_category_en','location_type','gender','primary_tags','north_south_vendor',
                              'east_west_vendor','north_south','east_west','vendor_cluster','cutomer_cluster','cluster_label_overall'])
reduce_mem_usage(train_df1)
reduce_mem_usage(test_df)
```

## **Step 3:-**
Model creation and training

```
import lightgbm as lgb 
model1 = lgb.LGBMClassifier(random_state=42,n_estimators=3000,silent=True,
                           scale_pos_weight=9,colsample_bylevel=0.8,subsample=0.8,learning_rate=0.05,
                           ) 
model2 = lgb.LGBMClassifier(random_state=42,n_estimators=1000,silent=True,
                           scale_pos_weight=9,colsample_bylevel=0.8,subsample=0.8,learning_rate=0.1,
                           ) 
model3 = lgb.LGBMClassifier(random_state=42,n_estimators=2000,silent=True,
                           scale_pos_weight=8.5,colsample_bylevel=0.8,subsample=0.8,learning_rate=0.07,
                           ) 
```

```
from sklearn.ensemble import VotingClassifier
clf = VotingClassifier(estimators=[('model1_lgb', model1), ('model2_lgb', model2), ('model3_lgb', model3)],
        voting='soft', weights=[2,1,1],
        )
clf.fit(train_df1,train_y)
```

**Final step** Creation of output files 

```
submission_df = pd.DataFrame(columns=['CID X LOC_NUM X VENDOR','target'])    
submission_df['CID X LOC_NUM X VENDOR'] = index_to_use_later
submission_df['target'] = clf.predict_proba(test_df)[:,1]
```

```
submission_df['target']=np.where(submission_df['target'] >0.57,1,0)
```
